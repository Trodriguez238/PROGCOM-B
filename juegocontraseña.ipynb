```python
import string
from collections import Counter

print("=== JUEGO DE LA CONTRASEÑA (Versión Simplificada) ===")
contrasena = input("Ingresa tu contraseña: ")

# Inicialización de reglas
tiene_mayus = False
tiene_minus = False
tiene_numero = False
tiene_simbolo = False
longitud_valida = False
longitud_maxima = False
no_contiene_password = True
no_contiene_admin = True
no_contiene_user = True
no_tiene_tres_seguidos = True
tiene_espacio = False
no_repite_caracteres_excesivos = True

# Conteo y revisión por carácter
conteo_caracteres = Counter(contrasena)
for i, c in enumerate(contrasena):
    if c.isupper(): tiene_mayus = True
    elif c.islower(): tiene_minus = True
    elif c.isdigit(): tiene_numero = True
    elif c in string.punctuation: tiene_simbolo = True
    if c == ' ': tiene_espacio = True

    if i >= 2:
        c1, c2 = contrasena[i-2], contrasena[i-1]
        # Secuencia ascendente
        if ord(c1)+1 == ord(c2) and ord(c2)+1 == ord(c):
            no_tiene_tres_seguidos = False
        # Repetición
        if c1 == c2 == c:
            no_tiene_tres_seguidos = False

# Revisión global
if len(contrasena) >= 8: longitud_valida = True
if len(contrasena) <= 50: longitud_maxima = True
if "password" in contrasena.lower(): no_contiene_password = False
if "admin" in contrasena.lower(): no_contiene_admin = False
if "user" in contrasena.lower(): no_contiene_user = False
for char, count in conteo_caracteres.items():
    if count > 4:
        no_repite_caracteres_excesivos = False
        break

# Verificación final
if (tiene_mayus and tiene_minus and tiene_numero and tiene_simbolo and
    longitud_valida and longitud_maxima and
    no_contiene_password and no_contiene_admin and no_contiene_user and
    no_tiene_tres_seguidos and tiene_espacio and no_repite_caracteres_excesivos):
    print("✅ Contraseña aceptada. Cumple las 12 reglas.")
else:
    print("❌ Contraseña rechazada. Faltan las siguientes reglas:")
    if not tiene_mayus: print("- Debe tener al menos una letra mayúscula")
    if not tiene_minus: print("- Debe tener al menos una letra minúscula")
    if not tiene_numero: print("- Debe tener al menos un número")
    if not tiene_simbolo: print("- Debe tener al menos un símbolo (!@#$%...)")
    if not longitud_valida: print("- Debe tener al menos 8 caracteres")
    if not longitud_maxima: print("- No debe superar los 50 caracteres")
    if not no_contiene_password: print("- No debe contener la palabra 'password'")
    if not no_contiene_admin: print("- No debe contener la palabra 'admin'")
    if not no_contiene_user: print("- No debe contener la palabra 'user'")
    if not no_tiene_tres_seguidos: print("- No debe tener 3 caracteres seguidos (ej: 'abc', '123', 'aaa')")
    if not tiene_espacio: print("- Debe contener al menos un espacio")
    if not no_repite_caracteres_excesivos: print("- No debe repetir ningún carácter más de 4 veces")
```
