```python
# Crear el archivo con frutas iniciales
file_path = 'frutas.txt'
frutas_iniciales = ["Manzana", "Pera", "Naranja"]
with open(file_path, 'w', encoding='utf-8') as f:
    for fruta in frutas_iniciales:
        f.write(fruta + "\n")

# Leer y mostrar numerado
with open(file_path, 'r', encoding='utf-8') as f:
    lineas = [line.strip() for line in f.readlines()]

for i, fruta in enumerate(lineas, start=1):
    print(f"{i}. {fruta}\n")

# Agregar nueva fruta al final sin borrar las anteriores
nueva_fruta = "Mango"
with open(file_path, 'a', encoding='utf-8') as f:
    f.write(nueva_fruta + "\n")

# Mostrar contenido actualizado
with open(file_path, 'r', encoding='utf-8') as f:
    lineas_actualizadas = [line.strip() for line in f.readlines()]

for i, fruta in enumerate(lineas_actualizadas, start=1):
    print(f"{i}. {fruta}\n")
```
