```python
from turtle import *
import colorsys
import random

speed(0)
bgcolor("black")
h = 0
for i in range(16):
    for j in range(18):
        c = colorsys.hsv_to_rgb(h, 1, 1)
        color(c)
        h += 0.005
        rt(90)
        circle(150 - j * 6, 90)
        lt(90)
        circle(150 - j * 6, 90)
        rt(180)
    circle(40, 24)

y = 500
for x in range(20):
    stick = Turtle()
    stick.speed(0)
    stick.color("#73FF00")
    screen = stick.getscreen()
    stick.left(15)
    stick.forward(x - 5)
    stick.right(90)
    stick.forward(y - 25)
    stick.right(90)
    stick.forward(x - 5)
    stick.right(90)
    stick.forward(y - 25)
    stick.right(90)
stick.goto(0, 0)

stick.color("black")

stick.goto(0, 42)
for z in range(50):
    stick.color("#E66553")
    stick.speed(0)
    stick.forward(z)
    stick.left(59)


grass = Turtle()
grass.hideturtle()
grass.speed(0)
grass.color("green")
grass.pensize(2)

for i in range(200):
    x = random.randint(-400, 400)
    y = random.randint(-300, -250)
    grass.penup()
    grass.goto(x, y)
    grass.pendown()
    grass.setheading(random.randint(80, 100))
    grass.forward(random.randint(20, 50))


# ---- Dibujar nubes ----
def draw_cloud(x, y, size=60):
    cloud = Turtle()
    cloud.hideturtle()
    cloud.speed(0)
    cloud.color("white")
    cloud.penup()
    cloud.goto(x, y)
    cloud.pendown()
    cloud.begin_fill()
    for _ in range(6):
        cloud.circle(size, 180)
        cloud.right(120)
    cloud.end_fill()

# Nubes más arriba
draw_cloud(-200, 260, 40)
draw_cloud(0, 280, 50)
draw_cloud(200, 240, 45)

done()
```
