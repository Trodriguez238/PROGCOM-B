```python
import streamlit as st
import yfinance as yf
import matplotlib.pyplot as plt
import csv
from datetime import datetime
import os

st.set_page_config(page_title="Asistente de InversiÃ³n", layout="centered")
st.title("ðŸ“Š Asistente de InversiÃ³n en Acciones")

# Entrada de datos
ticker = st.text_input("ðŸ”¤ Ticker de la acciÃ³n (ej. AAPL)").upper()
capital = st.number_input("ðŸ’° Capital disponible (USD)", min_value=0.0)
porcentaje = st.number_input("ðŸ“ˆ Porcentaje de ganancia deseado (%)", min_value=0.0)

# FunciÃ³n para guardar resultados
def guardar_resultado(datos):
    archivo = "inversiones.csv"
    existe = os.path.isfile(archivo)
    with open(archivo, mode="a", newline="") as f:
        writer = csv.writer(f)
        if not existe:
            writer.writerow([
                "Fecha", "Ticker", "Precio Actual", "Capital", "Porcentaje Deseado",
                "Acciones", "Precio Objetivo", "Ganancia Estimada", "RecomendaciÃ³n"
            ])
        writer.writerow(datos)

# EvaluaciÃ³n
if st.button("ðŸ“Œ Evaluar inversiÃ³n"):
    if not ticker:
        st.warning("Por favor ingresa un ticker vÃ¡lido.")
    else:
        datos = yf.Ticker(ticker).history(period="30d")
        if datos.empty:
            st.error("No se encontraron datos para ese ticker.")
        else:
            precio = datos["Close"].iloc[-1]
            acciones = int(capital // precio)
            if acciones == 0:
                st.error("No tienes suficiente capital para comprar una acciÃ³n.")
            else:
                precio_objetivo = precio * (1 + porcentaje / 100)
                ganancia = acciones * (precio_objetivo - precio)
                recomendacion = "âœ… Meta razonable" if porcentaje <= 15 else "âš ï¸ Meta ambiciosa, evalÃºa el riesgo"

                st.subheader("ðŸ“‹ Resultado de evaluaciÃ³n")
                st.write(f"**Precio actual:** ${precio:.2f}")
                st.write(f"**Acciones posibles:** {acciones}")
                st.write(f"**Precio objetivo:** ${precio_objetivo:.2f}")
                st.write(f"**Ganancia estimada:** ${ganancia:.2f}")
                st.write(f"**RecomendaciÃ³n:** {recomendacion}")

                guardar_resultado([
                    datetime.now().strftime("%Y-%m-%d %H:%M:%S"),
                    ticker, f"{precio:.2f}", f"{capital:.2f}", f"{porcentaje:.2f}",
                    acciones, f"{precio_objetivo:.2f}", f"{ganancia:.2f}", recomendacion
                ])

                # GrÃ¡fico
                st.subheader("ðŸ“‰ EvoluciÃ³n del precio (Ãºltimos 30 dÃ­as)")
                fig, ax = plt.subplots()
                ax.plot(datos.index, datos["Close"], marker='o', linestyle='-', color='blue')
                ax.set_title(f"Precio de cierre de {ticker}")
                ax.set_xlabel("Fecha")
                ax.set_ylabel("Precio (USD)")
                ax.grid(True)
                st.pyplot(fig)


```
