```python
import tkinter as tk
from tkinter import simpledialog, messagebox
import math
from scipy.stats import norm

# Fórmulas de Black-Scholes
def black_scholes(tipo, S, K, r, sigma, T):
    d1 = (math.log(S / K) + (r + 0.5 * sigma**2) * T) / (sigma * math.sqrt(T))
    d2 = d1 - sigma * math.sqrt(T)

    if tipo == "Call":
        precio = S * norm.cdf(d1) - K * math.exp(-r * T) * norm.cdf(d2)
    elif tipo == "Put":
        precio = K * math.exp(-r * T) * norm.cdf(-d2) - S * norm.cdf(-d1)
    else:
        raise ValueError("Tipo de opción inválido")
    return precio

# Interfaz gráfica
def calcular_opcion():
    tipo = simpledialog.askstring("Tipo de Opción", "Ingrese tipo (Call o Put):")
    if tipo is None or tipo.capitalize() not in ["Call", "Put"]:
        messagebox.showerror("Error", "Tipo inválido. Use 'Call' o 'Put'.")
        return

    try:
        S = float(simpledialog.askfloat("Precio del Activo", "Precio actual del activo (S):"))
        K = float(simpledialog.askfloat("Precio de Ejercicio", "Precio de ejercicio (K):"))
        r = float(simpledialog.askfloat("Tasa Libre de Riesgo", "Tasa anual (r) en decimal (ej. 0.05):"))
        sigma = float(simpledialog.askfloat("Volatilidad", "Volatilidad anual (σ) en decimal (ej. 0.2):"))
        T = float(simpledialog.askfloat("Tiempo", "Tiempo hasta vencimiento en años (T):"))

        precio = black_scholes(tipo.capitalize(), S, K, r, sigma, T)
        messagebox.showinfo("Resultado", f"Precio de la opción {tipo.capitalize()}: ${precio:,.4f}")

    except Exception as e:
        messagebox.showerror("Error", f"Entrada inválida: {e}")

# Menú principal
def iniciar_app():
    ventana = tk.Tk()
    ventana.title("Derivados Financieros - Opción Europea")
    ventana.geometry("400x250")

    tk.Label(ventana, text="Calculadora de Opciones (Black-Scholes)", font=("Arial", 14, "bold")).pack(pady=20)
    tk.Button(ventana, text="Calcular Opción Europea", command=calcular_opcion, width=30, height=2).pack(pady=10)
    tk.Button(ventana, text="Salir", command=ventana.quit, width=30, height=2).pack(pady=10)

    ventana.mainloop()

if __name__ == "__main__":
    iniciar_app()


```
