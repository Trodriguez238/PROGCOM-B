```python
#include <iostream>
using namespace std;

int main() {
    int numero;
    bool esPrimo = true;

    cout << "Ingrese un numero: ";
    cin >> numero;

    if (numero <= 1) {
        esPrimo = false;  // Los números menores o iguales a 1 no son primos
    } else {
        for (int i = 2; i <= numero / 2; i++) {
            if (numero % i == 0) {
                esPrimo = false;
                break; // Sale del ciclo porque ya encontró un divisor
            }
        }
    }

    if (esPrimo) {
        cout << numero << " es un numero primo." << endl;
    } else {
        cout << numero << " no es un numero primo." << endl;
    }

    return 0;
}
```
