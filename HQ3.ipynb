```python
import math

# --- Tablero inicial ---
board = [" " for _ in range(9)]

# --- Mostrar tablero actual ---
def print_board():
    print()
    for i in range(3):
        print(" | ".join(board[i*3:(i+1)*3]))
        if i < 2:
            print("---------")
    print()

# --- Mostrar gu칤a con posiciones ---
def print_position_guide():
    print("Posiciones del tablero:")
    guide = [str(i+1) for i in range(9)]
    for i in range(3):
        print(" | ".join(guide[i*3:(i+1)*3]))
        if i < 2:
            print("---------")
    print()

# --- Verificar ganador ---
def winner(b, player):
    win_positions = [
        [0, 1, 2],
        [3, 4, 5],
        [6, 7, 8],
        [0, 3, 6],
        [1, 4, 7],
        [2, 5, 8],
        [0, 4, 8],
        [2, 4, 6]
    ]
    return any(all(b[pos] == player for pos in combo) for combo in win_positions)

# --- Verificar si hay empate ---
def is_draw(b):
    return " " not in b

# --- Movimiento del jugador ---
def player_move():
    while True:
        move = input("Elige una posici칩n (1-9): ")
        if move.isdigit():
            move = int(move) - 1
            if 0 <= move < 9 and board[move] == " ":
                board[move] = "X"
                break
        print("Movimiento inv치lido. Intenta de nuevo.")

# --- Algoritmo minimax para la IA ---
def minimax(b, depth, is_maximizing):
    if winner(b, "O"):
        return 1
    if winner(b, "X"):
        return -1
    if is_draw(b):
        return 0

    if is_maximizing:
        best_score = -math.inf
        for i in range(9):
            if b[i] == " ":
                b[i] = "O"
                score = minimax(b, depth + 1, False)
                b[i] = " "
                best_score = max(score, best_score)
        return best_score
    else:
        best_score = math.inf
        for i in range(9):
            if b[i] == " ":
                b[i] = "X"
                score = minimax(b, depth + 1, True)
                b[i] = " "
                best_score = min(score, best_score)
        return best_score

# --- Movimiento de la IA ---
def ai_move():
    best_score = -math.inf
    move = None
    for i in range(9):
        if board[i] == " ":
            board[i] = "O"
            score = minimax(board, 0, False)
            board[i] = " "
            if score > best_score:
                best_score = score
                move = i
    board[move] = "O"

# --- Bucle principal ---
print("춰Bienvenido a Tic Tac Toe!")
print("T칰 eres X y la m치quina es O.")
print_position_guide()  # Mostrar gu칤a antes de jugar
print_board()

while True:
    player_move()
    print_board()
    if winner(board, "X"):
        print("춰Ganaste! 游꿀")
        break
    if is_draw(board):
        print("Empate 游뱋")
        break

    print("Turno de la m치quina...")
    ai_move()
    print_board()
    if winner(board, "O"):
        print("La m치quina gan칩 游땓")
        break
    if is_draw(board):
        print("Empate 游뱋")
        break

```
