```python
# Clase base
class PasoBase:
    def ejecutar(self):
        print("Inicio de la coreografÃ­a")

# Cada paso hereda del anterior
class PoseInicial(PasoBase):
    def ejecutar(self):
        super().ejecutar()
        print("estamos todos tiesos")

class MovimientoPiernas(PoseInicial):
    def ejecutar(self):
        super().ejecutar()
        print("todos hacemos paso cruzado con la pierna izquierda alfrente")
        print("todos alzamos la pierna derecha y alzamos brazos")
        print("todos hacemos paso cruzado con la pierna izquierda atras bajando brazos")
        print("todos hacemos paso cruzado alfrente alando brazos")

class PatadaCorta(MovimientoPiernas):
    def ejecutar(self):
        super().ejecutar()
        print("todos hacemos una patada hacia adelante alzando la pierna izquierda, con el brazo derecho doblado y extendiendo el brazo derecho hacia la izquierda")

class Pasito(PatadaCorta):
    def ejecutar(self):
        super().ejecutar()
        print("pasito a la derecha")
        print("pasito hacia la izquierda")

class PatadaPuno(Pasito):
    def ejecutar(self):
        super().ejecutar()
        print("patada")
        print("lanzar el puÃ±o")

class GirarCabeza(PatadaPuno):
    def ejecutar(self):
        super().ejecutar()
        print("cruzar las piernas con los brazos abiertos")
        print("girar la cabeza a la derecha")

class Paso7(GirarCabeza):
    def ejecutar(self):
        super().ejecutar()
        print("Pasito a la derecha")
        print("Pasito a la izquierda")

class Paso8(Paso7):
    def ejecutar(self):
        super().ejecutar()
        print("cierro las piernas")
        print("abro las piernas")

class Paso9(Paso8):
    def ejecutar(self):
        super().ejecutar()
        print("Levanto el brazo derecho y la pierna izquierda")

class PasitoIzq(Paso9):
    def ejecutar(self):
        super().ejecutar()
        print("saltico a la derecha hacia adelante")
        print("pasito a la izquierda hacia adelante")

class Meneadito(PasitoIzq):
    def ejecutar(self):
        super().ejecutar()
        print("pasito a la derecha")
        print("pasito a la izquierda")
        print("mueve cadera")

class Hulk(Meneadito):
    def ejecutar(self):
        super().ejecutar()
        print("sube y abre brazos")
        print("baja y cruza brazos")

# Clase que ejecuta toda la coreografÃ­a
class CoreografiaCompleta(Hulk):
    def ejecutar(self):
        print("ðŸŽ¶ Iniciando la coreografÃ­a completa ðŸŽ¶")
        super().ejecutar()
        print("ðŸŽ¶ Fin de la coreografÃ­a ðŸŽ¶")
```
