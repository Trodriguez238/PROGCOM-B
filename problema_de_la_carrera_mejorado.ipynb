```python
import tkinter as tk
import random

def evaluar_inversion():
    # Borrar gr√°fica anterior
    canvas.delete("all")

    try:
        porcentaje = float(entry_porcentaje.get())
        precio_actual = float(entry_precio.get())
    except ValueError:
        resultado_text.set("‚ùå Ingresa valores num√©ricos v√°lidos.")
        return

    # Evaluaci√≥n de rentabilidad
    if porcentaje <= 15:
        resultado_text.set("‚úÖ La inversi√≥n es rentable (meta razonable).")
    else:
        resultado_text.set("‚ö†Ô∏è Meta alta. Inversi√≥n arriesgada.")

    # --- Datos simulados para la gr√°fica (30 d√≠as) ---
    precios = []
    for i in range(30):
        variaci√≥n = random.uniform(-0.03, 0.03)   # variaci√≥n simulada ¬±3%
        precio_actual += precio_actual * variaci√≥n
        precios.append(precio_actual)

    # --- Dibujar la gr√°fica ---
    dibujar_grafico(precios)


def dibujar_grafico(precios):
    # Dimensiones del canvas
    ancho = 400
    alto = 200
    margen = 20

    max_p = max(precios)
    min_p = min(precios)

    # Evitar divisi√≥n por cero si todos los precios son iguales
    rango = max_p - min_p if max_p != min_p else 1

    # Ejes
    canvas.create_line(margen, alto - margen, ancho - margen, alto - margen, width=2)  # eje X
    canvas.create_line(margen, margen, margen, alto - margen, width=2)  # eje Y

    # L√≠nea de la gr√°fica
    for i in range(len(precios) - 1):
        x1 = margen + (i * (ancho - 2 * margen) / 29)
        y1 = alto - margen - ((precios[i] - min_p) * (alto - 2 * margen) / rango)

        x2 = margen + ((i + 1) * (ancho - 2 * margen) / 29)
        y2 = alto - margen - ((precios[i + 1] - min_p) * (alto - 2 * margen) / rango)

        canvas.create_line(x1, y1, x2, y2, fill="blue", width=2)

    canvas.create_text(ancho/2, 10, text="üìâ Evoluci√≥n simulada de la acci√≥n", font=("Arial", 10))


# --- Interfaz gr√°fica ---
ventana = tk.Tk()
ventana.title("üìä Asistente de Inversi√≥n en Acciones")
ventana.geometry("500x550")

# Entradas
tk.Label(ventana, text="üî§ Ticker de la acci√≥n (ej. AAPL)").pack(pady=5)
entry_ticker = tk.Entry(ventana)
entry_ticker.pack()

tk.Label(ventana, text="üíµ Precio actual de la acci√≥n (USD)").pack(pady=5)
entry_precio = tk.Entry(ventana)
entry_precio.pack()

tk.Label(ventana, text="üìà Porcentaje de ganancia deseado (%)").pack(pady=5)
entry_porcentaje = tk.Entry(ventana)
entry_porcentaje.pack()

# Bot√≥n
tk.Button(ventana, text="üìå Evaluar inversi√≥n", command=evaluar_inversion).pack(pady=10)

# Resultado
resultado_text = tk.StringVar()
tk.Label(ventana, textvariable=resultado_text, justify="left").pack(pady=10)

# Canvas para gr√°fica
canvas = tk.Canvas(ventana, width=400, height=200, bg="white")
canvas.pack(pady=10)

ventana.mainloop()
```
