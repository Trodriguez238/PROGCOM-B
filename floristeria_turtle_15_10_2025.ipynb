```python
import turtle

# Configuración inicial
screen = turtle.Screen()
screen.title("Casita decorada con flores traseras y puerta centrada")
screen.bgcolor("lightyellow")
t = turtle.Turtle()
t.speed(0)

# Funciones básicas
def dibujar_rectangulo(t, ancho, alto, color):
    t.fillcolor(color)
    t.begin_fill()
    for _ in range(2):
        t.forward(ancho)
        t.left(90)
        t.forward(alto)
        t.left(90)
    t.end_fill()

def dibujar_triangulo(t, base, altura, color):
    t.fillcolor(color)
    t.begin_fill()
    t.forward(base)
    t.left(135)
    t.forward(altura)
    t.left(90)
    t.forward(altura)
    t.left(135)
    t.end_fill()

def dibujar_rayas_verticales(t, x, y, ancho, alto, color, cantidad):
    espacio = ancho // cantidad
    t.pensize(3)  # Grosor aumentado
    for i in range(cantidad):
        t.penup()
        t.goto(x + i * espacio, y)
        t.setheading(90)
        t.pendown()
        t.pencolor(color)
        t.forward(alto)
    t.pensize(1)  # Restaurar grosor normal

def dibujar_circulo(t, x, y, color, radio):
    t.penup()
    t.goto(x, y)
    t.pendown()
    t.color(color)
    t.begin_fill()
    t.circle(radio)
    t.end_fill()

def dibujar_tallo(t, x, y, altura):
    t.penup()
    t.goto(x, y)
    t.setheading(90)
    t.pendown()
    t.pencolor("green")
    t.pensize(2)
    t.forward(altura)
    t.pensize(1)

def dibujar_techo_blanco(x, y, base, altura):
    t.penup()
    t.goto(x, y)
    t.setheading(0)
    t.pendown()
    dibujar_triangulo(t, base, altura, "white")

# Vista frontal
def vista_frontal(x, y):
    t.penup()
    t.goto(x, y)
    t.setheading(0)
    t.pendown()
    dibujar_rectangulo(t, 100, 100, "white")
    dibujar_techo_blanco(x, y + 100, 100, 70)
    dibujar_rayas_verticales(t, x, y, 100, 100, "orange", 10)

    # Puerta centrada en la parte inferior
    t.penup()
    t.goto(x + 35, y)
    t.setheading(0)
    t.pendown()
    dibujar_rectangulo(t, 30, 50, "sienna")

    # Flor decorativa encima de la puerta
    dibujar_circulo(t, x + 50, y + 70, "yellow", 10)
    dibujar_circulo(t, x + 50, y + 70, "red", 4)

# Vista trasera con tres flores con tallo
def vista_trasera(x, y):
    t.penup()
    t.goto(x, y)
    t.setheading(0)
    t.pendown()
    dibujar_rectangulo(t, 100, 100, "white")
    dibujar_techo_blanco(x, y + 100, 100, 70)
    dibujar_rayas_verticales(t, x, y, 100, 100, "orange", 10)

    # Tres flores con tallo
    posiciones = [x + 20, x + 50, x + 80]
    for px in posiciones:
        dibujar_tallo(t, px, y + 20, 20)
        dibujar_circulo(t, px, y + 40, "yellow", 15)

# Vista lateral izquierda con ventana
def vista_lateral_izquierda(x, y):
    t.penup()
    t.goto(x, y)
    t.setheading(0)
    t.pendown()
    dibujar_rectangulo(t, 60, 100, "white")
    dibujar_techo_blanco(x, y + 100, 60, 42)
    dibujar_rayas_verticales(t, x, y, 60, 100, "orange", 6)

    # Ventana azul
    t.penup()
    t.goto(x + 20, y + 60)
    t.setheading(0)
    t.pendown()
    dibujar_rectangulo(t, 20, 20, "skyblue")

# Vista lateral derecha sin ventana
def vista_lateral_derecha(x, y):
    t.penup()
    t.goto(x, y)
    t.setheading(0)
    t.pendown()
    dibujar_rectangulo(t, 60, 100, "white")
    dibujar_techo_blanco(x, y + 100, 60, 42)
    dibujar_rayas_verticales(t, x, y, 60, 100, "orange", 6)

# Vista superior
def vista_superior(x, y):
    t.penup()
    t.goto(x, y)
    t.setheading(0)
    t.pendown()
    dibujar_rectangulo(t, 100, 80, "white")

# Dibujo completo
vista_frontal(-250, 100)
vista_trasera(-80, 100)
vista_lateral_izquierda(100, 100)
vista_lateral_derecha(220, 100)
vista_superior(-50, -150)

t.hideturtle()
screen.mainloop()
```
