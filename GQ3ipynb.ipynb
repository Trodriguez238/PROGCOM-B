```python
# -*- coding: utf-8 -*-
"""
Análisis Finencial Simple
Hecho por: Estudiante de Segundo Semestre
Fecha: [Fecha actual]
"""


# IMPORTAR LIBRERÍAS (las que hemos visto en clase)
import math  # para cálculos matemáticos
import statistics  # para promedios y estadísticas básicas

# FUNCIONES QUE HE APRENDIDO EN CLASE

def mostrar_bienvenida():
    """Función para mostrar mensaje de bienvenida"""
    print("=" * 50)
    print("   ANÁLISIS FINANCIERO BÁSICO")

    print("=" * 50)
    print()

def calcular_promedio(lista_numeros):
    """Calcula el promedio de una lista de números"""
    if len(lista_numeros) == 0:
        return 0
    return sum(lista_numeros) / len(lista_numeros)

def encontrar_maximo(lista_numeros):
    """Encuentra el valor máximo en una lista"""
    if len(lista_numeros) == 0:
        return 0
    return max(lista_numeros)

def encontrar_minimo(lista_numeros):
    """Encuentra el valor mínimo en una lista"""
    if len(lista_numeros) == 0:
        return 0
    return min(lista_numeros)

def calcular_interes_simple(capital, tasa, tiempo):
    """Calcula interés simple: I = P * r * t"""
    interes = capital * (tasa / 100) * tiempo
    return interes

def calcular_interes_compuesto(capital, tasa, tiempo):
    """Calcula interés compuesto: A = P(1 + r)^t"""
    monto = capital * (1 + tasa/100) ** tiempo
    return monto

def analizar_ventas_mensuales():
    """Analiza las ventas mensuales de una empresa"""
    print("\n--- ANÁLISIS DE VENTAS MENSUALES ---")

    # Pedir datos al usuario
    meses = ["Enero", "Febrero", "Marzo", "Abril", "Mayo", "Junio",
             "Julio", "Agosto", "Septiembre", "Octubre", "Noviembre", "Diciembre"]

    ventas = []
    for mes in meses:
        while True:
            try:
                venta = float(input(f"Ingrese las ventas de {mes}: $"))
                ventas.append(venta)
                break
            except ValueError:
                print("¡Error! Por favor ingrese un número válido.")

    # Cálculos básicos
    total_ventas = sum(ventas)
    promedio_ventas = calcular_promedio(ventas)
    max_venta = encontrar_maximo(ventas)
    min_venta = encontrar_minimo(ventas)

    # Mostrar resultados
    print("\n" + "="*40)
    print("RESULTADOS DEL ANÁLISIS DE VENTAS:")
    print("="*40)
    print(f"Total de ventas anual: ${total_ventas:,.2f}")
    print(f"Promedio mensual: ${promedio_ventas:,.2f}")
    print(f"Mejor mes: ${max_venta:,.2f}")
    print(f"Peor mes: ${min_venta:,.2f}")

    # Mes con mejores y peores ventas
    mejor_mes = meses[ventas.index(max_venta)]
    peor_mes = meses[ventas.index(min_venta)]
    print(f"Mes con mejores ventas: {mejor_mes}")
    print(f"Mes con peores ventas: {peor_mes}")

def analizar_gastos():
    """Analiza los gastos de una persona o empresa"""
    print("\n--- ANÁLISIS DE GASTOS ---")

    categorias = ["Alimentación", "Transporte", "Vivienda", "Entretenimiento", "Otros"]
    gastos = []

    for categoria in categorias:
        while True:
            try:
                gasto = float(input(f"Ingrese gastos en {categoria}: $"))
                gastos.append(gasto)
                break
            except ValueError:
                print("¡Error! Por favor ingrese un número válido.")

    total_gastos = sum(gastos)

    print("\n" + "="*40)
    print("DISTRIBUCIÓN DE GASTOS:")
    print("="*40)

    for i in range(len(categorias)):
        porcentaje = (gastos[i] / total_gastos) * 100 if total_gastos > 0 else 0
        print(f"{categorias[i]}: ${gastos[i]:,.2f} ({porcentaje:.1f}%)")

    print(f"\nTotal de gastos: ${total_gastos:,.2f}")

    # Recomendación básica
    mayor_gasto = categorias[gastos.index(max(gastos))]
    print(f"Área con mayor gasto: {mayor_gasto}")

def calcular_ahorro_objetivo():
    """Calcula cuánto tiempo tomará alcanzar un objetivo de ahorro"""
    print("\n--- CALCULADORA DE AHORRO ---")

    try:
        ahorro_actual = float(input("Ingrese su ahorro actual: $"))
        objetivo = float(input("Ingrese su objetivo de ahorro: $"))
        ahorro_mensual = float(input("Ingrese cuánto puede ahorrar por mes: $"))

        if ahorro_mensual <= 0:
            print("El ahorro mensual debe ser mayor a cero.")
            return

        falta_ahorrar = objetivo - ahorro_actual

        if falta_ahorrar <= 0:
            print("¡Felicidades! Ya alcanzaste tu objetivo.")
            return

        meses_necesarios = math.ceil(falta_ahorrar / ahorro_mensual)
        años = meses_necesarios // 12
        meses = meses_necesarios % 12

        print(f"\nPara alcanzar tu objetivo de ${objetivo:,.2f}:")
        print(f"Faltan: ${falta_ahorrar:,.2f}")
        print(f"Tiempo estimado: {años} años y {meses} meses")

    except ValueError:
        print("¡Error! Por favor ingrese números válidos.")

def menu_principal():
    """Menú principal del programa"""
    while True:
        print("\n" + "="*50)
        print("MENÚ PRINCIPAL - ANÁLISIS FINANCIERO")
        print("="*50)
        print("1. Análisis de Ventas Mensuales")
        print("2. Análisis de Gastos")
        print("3. Calculadora de Ahorro")
        print("4. Calculadora de Interés Simple")
        print("5. Calculadora de Interés Compuesto")
        print("6. Salir")

        opcion = input("\nSeleccione una opción (1-6): ")

        if opcion == "1":
            analizar_ventas_mensuales()
        elif opcion == "2":
            analizar_gastos()
        elif opcion == "3":
            calcular_ahorro_objetivo()
        elif opcion == "4":
            print("\n--- INTERÉS SIMPLE ---")
            try:
                capital = float(input("Capital inicial: $"))
                tasa = float(input("Tasa de interés anual (%): "))
                tiempo = float(input("Tiempo (años): "))
                interes = calcular_interes_simple(capital, tasa, tiempo)
                print(f"Interés generado: ${interes:,.2f}")
                print(f"Total final: ${capital + interes:,.2f}")
            except ValueError:
                print("¡Error! Ingrese valores numéricos válidos.")
        elif opcion == "5":
            print("\n--- INTERÉS COMPUESTO ---")
            try:
                capital = float(input("Capital inicial: $"))
                tasa = float(input("Tasa de interés anual (%): "))
                tiempo = float(input("Tiempo (años): "))
                monto = calcular_interes_compuesto(capital, tasa, tiempo)
                print(f"Monto final: ${monto:,.2f}")
                print(f"Interés ganado: ${monto - capital:,.2f}")
            except ValueError:
                print("¡Error! Ingrese valores numéricos válidos.")
        elif opcion == "6":
            print("\n¡Gracias por usar el programa! Hasta pronto.")
            break
        else:
            print("Opción no válida. Por favor seleccione 1-6.")

#
# PROGRAMA PRINCIPAL
#
if __name__ == "__main__":
    mostrar_bienvenida()
    menu_principal()
```
